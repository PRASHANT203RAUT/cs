```python
X=df['WINS']
Y=df['LOSSES']
```

```python
def estfun(X,Y):
  n =np.size(X) 
  m_x =np.mean(X)
  m_y = np.mean(Y) 
  SS_xy =np.sum(Y*X)-n*m_y*m_x 
  SS_xx = np.sum(X*X) - n*m_x*m_x 
  b1 = SS_xy/SS_xx
  b0 = m_y -b1*m_x
  return(B0,B1)
```

```python
B = estfun(X,Y)
print("Beta 0:",B[0],"\n Beta 1",B[1])
y=B[0]+B[1]
print("y =",B[0],"+(",B[1],")X")
```
